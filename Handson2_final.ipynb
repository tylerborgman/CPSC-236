## 3.1 Letter Grade Converter
Create a program that converts number grades to letter grades.

### Console:
```powershell
Letter Grade Converter

Enter numerical grade: 90 
Letter grade: A

Continue? (y/n): y

Enter numerical grade: 88 
Letter grade: A

Continue? (y/n): y

Enter numerical grade: 80

Enter numerical grade: 67 
Letter grade: C

Continue? (y/n): y

Enter numerical grade: 59 
Letter grade: F

Continue? (y/n): n

```

### Specifications:
- Assume that the user will enter valid integers for the grades.
- The program should continue only if the user enters “y” or “Y” to continue.
- The grading criteria is as follows:

| Letter |	Range |
|---|-------|
| A |	88-100 |
| B |	80-87 |
| C |	67-79 |
| D |	60-66 |
| F |	<60 |

````python
##conditonals to get values
print("Letter Grade Converter")
print()
##to continue loop
choice = "y"
while choice.lower() == "y":

    grade = int(input("Enter Numerical Grade: "))
    if grade >= 88 and grade <= 100:
        print("Letter grade: A")
    elif grade >= 80 and grade <= 87:
        print("Letter grade: B")
    elif grade >= 67 and grade <= 79:
        print("Letter grade: C")
    elif grade >= 60 and grade <= 66:
        print("Letter grade: D")
    elif grade < 60:
        print("Letter grade: F")
    print()

##to continue loop

    choice= input("Continue? (y/n): ")
    if choice == "n":
        print("Bye")
        break
````

## 3.3 - Change Calculator
Create a program that calculates the coins needed to make change for the specified number of cents.

### Console
```powershell
Change Calculator

Enter number of cents (0-99): 99

Quarters: 3
Dimes:	2
Nickels: 0
Pennies: 4

Continue? (y/n): y

Enter number of cents (0-99): 55

Quarters: 2
Dimes:	0
Nickels: 1
Pennies: 0

Continue? (y/n): n 

Bye!
```

### Specifications
- The program should display the minimum number of quarters, dimes, nickels, and pennies that one needs to make up the specified number of cents.
- Assume that the user will enter a valid integer for the number of cents.
- The program should continue only if the user enters “y” or “Y” to continue.


````python
##function to calculate change
def calculate_change(cents):
    quarters = cents // 25
    cents %= 25
    
    dimes = cents // 10
    cents %= 10
    
    nickels = cents // 5
    cents %= 5
    
    pennies = cents
    
    print(f"Quarters: {quarters}")
    print(f"Dimes:  {dimes}")
    print(f"Nickels: {nickels}")
    print(f"Pennies: {pennies}")

    
def main():
    while True:
        try:
            cents = int(input("Enter number of cents (0-99): "))
            print()
            if 0 <= cents <= 99:
                calculate_change(cents)
            else:
                print("Please enter a valid integer between 0 and 99.")
                print()
                continue
        except ValueError:
            print("Please enter a valid integer.")
            print()
            continue
        
        user_input = input("Continue? (y/n): ")
        print()
        if user_input.lower() != 'y':
            print("Bye!")
            print()
            break

if __name__ == "__main__":
    print("Change Calculator")
    print()
    main()

````

## 3.4 - Table of Powers
Create a program that displays a table of squares and cubes for the specified range of numbers.

### Console
```powershell
Table of Powers

Start number: 90
Stop number: 100

Number	Squared	Cubed
======	=======	=====
90      8100	729000
91      8281	753571
92      8464	778688
93      8649	804357
94      8836	830584
95      9025	857375
96      9216	884736
97      9409	912673
98      9604	941192
99      9801	970299
100     10000	1000000
```

### Specifications
- The formulas for calculating squares and cubes are:

    `square = x ** 2`
    
    `cube = x ** 3`

- Use tabs to align the columns.
- Assume that the user will enter valid integers.
- Make sure the user enters a start integer that’s less than the stop integer. If the user enters a start integer that’s greater than the stop integer, display an error message and give the user a chance to enter the integers again.


````python
##function for start / stop numbers calculations
def calculate_powers(start, stop):
    print("Number\tSquared\tCubed")
    print("======\t=======\t=====")
    
    for num in range(start, stop + 1):
        squared = num ** 2
        cubed = num ** 3
        print(f"{num}\t{squared}\t{cubed}")

#main function
def main():
    while True:
        try:
            start = int(input("Start number: "))
            stop = int(input("Stop number: "))
            print()
            
            if start >= stop:
                print("Error: Start number must be less than the stop number.")
                continue
            else:
                calculate_powers(start, stop)
        except ValueError:
            print("Please enter valid integers.")
            continue
        
        break

if __name__ == "__main__":
    print("Table of Powers")
    print()
    main()
````
